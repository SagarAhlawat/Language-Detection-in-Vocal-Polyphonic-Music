```python
import os
import random
import math
from pydub import AudioSegment
DATASET_PATH = "python_jupyter/DATA_DOWN/B_A"
save_path = 'python_jupyter/DATA_DOWN/B'
#print(os.listdir(DATASET_PATH))
for filename in os.listdir(DATASET_PATH):
    try:
        audio = AudioSegment.from_wav(os.path.join(DATASET_PATH, filename))
        endt=math.floor(audio.duration_seconds)
        
        list_of_timestamps =random.sample(range(0, endt), 30)
        
        for idx,i in enumerate(list_of_timestamps):
            if idx == len(list_of_timestamps):
                break
            start=(i-10)*1000;
            end = i * 1000
            audio_chunk=audio[start:end]
            fname=str(filename[:-3])+str(idx)+'.wav'
            completeName = os.path.join(save_path, fname)
            audio_chunk.export( completeName.format(end), format="wav")
            print ("exporting", audio_chunk)
            
        
    except:
        print("an error occured")
```

```python
from pydub import AudioSegment 
from pydub.utils import make_chunks
import os

```

```python


def process_sudio(file_name):
    myaudio = AudioSegment.from_file(file_name, format="mp4") 
    chunk_length_ms = 8000 # pydub calculates in millisec 
    chunks = make_chunks(myaudio,chunk_length_ms) #Make chunks of one sec 
    for i, chunk in enumerate(chunks): 
        chunk_name = file_name + "_{0}.wav".format(i) 
        print ("exporting", chunk_name) 
        chunk.export(chunk_name, format="wav")


```

```python
import os
DATASET_PATH = "HINDI_audio"
#print(os.listdir(DATASET_PATH))
for filename in os.listdir(DATASET_PATH):
    process_sudio(os.path.join("HINDI_audio", filename))
    
 

```
